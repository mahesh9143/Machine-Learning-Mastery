```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv("D:\dataScience\Feature-Engineering-master\IRIS.csv")
```

```python
df.head()
```

```python
df['species'].unique()
```

```python
df_SETOSA = df.loc[df['species'] == "Iris-setosa"]
df_versicolor = df.loc[df['species'] == "Iris-versicolor"]
df_virginica = df.loc[df['species'] == "Iris-virginica"]
```

```python
df_SETOSA.head()
```

```python
df_versicolor.head()
```

# Uni variate analysis




```python
len(df_SETOSA['sepal_length'])
```

```python
plt.plot(df_SETOSA['sepal_length'] , np.zeros_like(df_SETOSA['sepal_length']),marker = 'o', c = 'red')
plt.plot(df_versicolor['sepal_length'] , np.zeros_like(df_versicolor['sepal_length']),marker = 'o', c = 'yellow')
plt.plot(df_virginica['sepal_length'] , np.zeros_like(df_virginica['sepal_length']),marker = '*', c = 'green')
plt.show()
```

# Bi Variate Analysis

```python
sns.FacetGrid(df,hue = 'species',size = 7).map(plt.scatter,'sepal_length','sepal_width').add_legend()
plt.show()
```

```python
sns.FacetGrid(df,hue = 'species',height = 5).map(plt.scatter,'petal_length','petal_width').add_legend()
plt.show()
```

# Mutli Variate Analysis

```python
sns.pairplot(df,hue = 'species',size = 5).add_legend()
plt.show()
```
